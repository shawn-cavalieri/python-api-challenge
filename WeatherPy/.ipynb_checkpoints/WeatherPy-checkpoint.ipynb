3 Takeaways:

1) While we already knew this, we can clearly see that the highest temperatures are at 0 degrees latitude (the equator). We can see very cold temps in positive latitudes (northern hemisphere) as we have reached winter in the north.

2) It does not appear that there are any notable correlations between latitude and cloudiness.

3) Northern Hemisphere citites were much more likely to have a higher humidity % than southern hemisphere cities, however this is a rather weak correlation.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python

# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Perform API Calls
# Perform a weather check on each city using a series of successive API calls.
# Include a print log of each city as it'sbeing processed (with the city number and city name).

# url for api call

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"

query_url = f"{url}appid={weather_api_key}&units={units}&q="


# counter for cities

counter = 1

# create lists for holding items

city_name = []
cloud = []
country = []
date = []
humidity = []
lat = []
long = []
max_temp = []
wind_speed = []

# create a for loop to go through each city and collect items

print("-------------")
print("Beginning Data Retrieval")
print("-------------")


for city in cities:
    try:
        data = requests.get(query_url + city).json()
        city_name.append(data['name'])
        cloud.append(data["clouds"]["all"])
        country.append(data["sys"]["country"])
        date.append(data["dt"])
        humidity.append(data["main"]["humidity"])
        lat.append(data["coord"]["lat"])
        long.append(data["coord"]["lon"])
        max_temp.append(data["main"]["temp_max"])
        wind_speed.append(data["wind"]["speed"])
        print(f"Processing Record {counter} of {len(cities)}. | {city}")
        counter = counter + 1
       
    except KeyError:
        print(f"City not found. Skipping {city}.")
        counter = counter + 1
        time.sleep(0.1)
        
```

```python
# save the data into a dataframe and display head

city_data = pd.DataFrame({"City": city_name,
                         "Cloudiness": cloud,
                         "Country": country,
                         "Date": date,
                         "Humidity": humidity,
                         "Latitude": lat,
                         "Longitude": long,
                         "Max Temperature": max_temp,
                         "Wind Speed": wind_speed})

city_data.head()

# export city data into a csv

pd.DataFrame.to_csv(city_data, 'city_dataframe.csv')
```

```python
# Plotting the Data
# Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
# Save the plotted figures as .pngs.

# latitude vs. temperature plot

lat = city_data["Latitude"]
temp = city_data["Max Temperature"]

plt.scatter(lat, temp)

plt.title("City Latitude vs. Max Temperature (1/07/2020)")
plt.xlabel("City Latitude")
plt.ylabel("Max City Temperature (F)")

plt.savefig('latvstemp.png')

plt.show()
```

```python
# latitude vs. cloudiness plot

lat = city_data["Latitude"]
cloud = city_data["Cloudiness"]

plt.scatter(lat, cloud)

plt.title("City Latitude vs. Cloudiness (1/07/2020)")
plt.xlabel("City Latitude")
plt.ylabel("Cloudiness (%)")

plt.savefig("latvscloud.png")

plt.show()
```

```python
# latitude vs. wind speed plot

lat = city_data["Latitude"]
wind = city_data["Wind Speed"]

plt.scatter(lat, wind)

plt.title("City Latitude vs. Wind Speed (1/07/2020)")
plt.xlabel("City Latitude")
plt.ylabel("Wind Speed (MPH)")

plt.savefig("latvswind.png")

plt.show()
```

```python
# Linear Regression

# create a function to seperate Northern vs. Southern Hemispheres

north_hemi = city_data.loc[(city_data["Latitude"] > 0)]

north_hemi

south_hemi = city_data.loc[(city_data["Latitude"] <= 0)]

south_hemi
```

```python
# northern hemisphere max temp vs. lat regression

## Convert to floats and store Latitude and Max Temp as x and y values
x_values = north_hemi["Latitude"].astype('float')
y_values = north_hemi["Max Temperature"].astype('float')

# Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot scatter plot
plt.scatter(x_values,y_values)

# Plot regression line
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

# Label plot
plt.title("Northern Hemisphere Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature")

# Print r square value
print(f"The R squared is : {rvalue}")

# Show plot
plt.show()
```
